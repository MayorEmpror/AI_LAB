```python



class Enviornment:
    def __init__(self, heat_level='High'):
            self._heat_level = heat_level
    def get_percept(self):
        return 'Hot' if self._heat_level == 'High' else 'Cool'  
    
class SimpleReflexAgent:
        def __init__(self): pass
        def act(self, percept):
            if(percept == 'Hot'): return 'Pull hand away, you touched the hot object'
            else : return 'You have not touched any hot object, No need to pull you hand away'

def run_agent(agent, enviornment):
            percept = enviornment.get_percept()
            action = agent.act(percept)
            print(f"Precept : {percept}, Action: {action}")


agent = SimpleReflexAgent()
env = Enviornment(heat_level='Low')
run_agent(agent, env)

```

```python


class SimpleRelfexAgent:
    def __init__(self, envmodel):
        self.position = 0
        self.environment_model = envmodel

    def act(self, percept):
        if(percept == 'Clean'):
            return "room is already clean"
        else:
            self.environment_model[self.position] = 'Clean'
            return "room coordinate has been cleaned"
    def move(self):
        if(self.position < 8): #since there are only 9 places to be
            self.position += 1 #move forward on the grid
        return self.position

    def updateModel(self, position, percept):
        self.environment_model[position] = percept

    def get_model(self):
        return self.environment_model

class Enviornment:
    def __init__(self):
         self.grid = ['Clean', 'Dirty', 'Clean',
                     'Clean', 'Dirty', 'Dirty',
                     'Clean', 'Clean', 'Clean']
    def get_percept(self, position):
            return self.grid[position]
    def clean_room(self, position):
            self.grid[position] = 'Clean'
    def display_grid(self, agent_position):
        print("\n Current grid State: ")
        grid_with_agent = self.grid[:]
        grid_with_agent[agent_position] = '😈'
        for i in range(0,9,3): # start at 0, go to 9 with a step of 3
            print(" | ".join(grid_with_agent[i:i + 3]))
        print()


def run_agent(agent, enviornment, steps):
        for step in range(steps):
            percept = enviornment.get_percept(agent.position)
            action = agent.act(percept)
            print(f"Step {step + 1}: Position {agent.position}: Action {action}: Percept : {percept}")

            agent.updateModel(agent.position, percept)
            enviornment.display_grid(agent.position)

            if percept == 'Dirty':
                enviornment.clean_room(agent.position)
            agent.move()

agent = SimpleReflexAgent()
env = Enviornment()
run_agent(agent, env,9)


```

```python
class GoalBaseAgent:
        def __init__(self):
                self.goal = 'Clean'
        def formulate_goal(self, percept):
                if percept == 'Dirty':
                    self.goal = 'Clean'
                else:
                    self.goal = 'No action needed'
        def act(self, percept):
                self.formulate_goal(percept)
                if(self.goal =='Clean'):
                    return 'Clean the room'
                else:
                    return 'Room is clean'

class Enviornment:
    def __init__(self, state='Dirty'):
            self.state = state
    def get_percept(self):
            return self.state

    def clean_room(self):
            self.state = 'Clean'

def run_agent(agent, enviornment, steps):
         for step in range(steps):
                percept = enviornment.get_percept()
                action = agent.act(percept)
                print(f"Step {step + 1}: Percept - {percept}, Action - {action}")
                if percept == 'Dirty':
                    enviornment.clean_room()

agent = GoalBaseAgent()
env = Enviornment() 

run_agent(agent, env, 9)

```

```python
import random

class LearningBasedAgent:

    def __init__(self, actions):
        self.Q = {}
        self.actions = actions
        self.alpha = 0.1
        self.gamma = 0.9
        self.elipson = 0.1 
    
    def get_Q_value(self, state, action):
        return self.Q.get((state, action), 0.0) # here 0.0 is a fallback value, if the get_Q_value funciton doeesnt find any thing related to the state action pair
    
    def select_action(self, state):
            if random.uniform(0, 1) < self.elipson:
                 return random.choice(self.actions)
            else:
                return max(self.actions, key=lambda a: self.get_Q_value(state, a))

    


    
```
