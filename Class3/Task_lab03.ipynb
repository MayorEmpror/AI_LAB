```python
import random

class SecurityEnvironment:
    def __init__(self):
        self.components = {chr(i): random.choice(["Safe", "Vulnerable"])
                           for i in range(ord('A'), ord('J'))}

    def display(self, title):
        print(f"\n{title}")
        for c, s in self.components.items():
            print(f"{c}: {s}")


class SecurityAgent:
    def scan_and_patch(self, env):
        to_patch = []

        print("\n Scanning System:")
        for c, state in env.components.items():
            if state == "Vulnerable":
                print(f" {c} vulnerable")
                to_patch.append(c)
            else:
                print(f" {c} secure")

        print("\n Patching:")
        for c in to_patch:
            env.components[c] = "Safe"
            print(f"{c} patched")


env = SecurityEnvironment()
agent = SecurityAgent()

env.display("Initial State")
agent.scan_and_patch(env)
env.display("Final State")
```

```python
class LoadEnvironment:
    def __init__(self):
        self.servers = {
            "S1": "Overloaded",
            "S2": "Underloaded",
            "S3": "Balanced",
            "S4": "Overloaded",
            "S5": "Underloaded"
        }

    def display(self):
        for s, l in self.servers.items():
            print(f"{s}: {l}")


class LoadBalancerAgent:
    def balance(self, env):
        overloaded = [s for s in env.servers if env.servers[s] == "Overloaded"]
        underloaded = [s for s in env.servers if env.servers[s] == "Underloaded"]

        for o, u in zip(overloaded, underloaded):
            print(f"⚖ Moving load from {o} → {u}")
            env.servers[o] = "Balanced"
            env.servers[u] = "Balanced"


env = LoadEnvironment()
agent = LoadBalancerAgent()

print("\nInitial Load:")
env.display()

agent.balance(env)

print("\nAfter Balancing:")
env.display()
```

```python
class BackupEnvironment:
    def __init__(self):
        self.tasks = {
            "B1": "Completed",
            "B2": "Failed",
            "B3": "Completed",
            "B4": "Failed"
        }

    def display(self):
        for t, s in self.tasks.items():
            print(f"{t}: {s}")


class BackupAgent:
    def retry_failed(self, env):
        for task in env.tasks:
            if env.tasks[task] == "Failed":
                print(f" Retrying {task}")
                env.tasks[task] = "Completed"


env = BackupEnvironment()
agent = BackupAgent()

print("\nInitial Backups:")
env.display()

agent.retry_failed(env)

print("\nFinal Backups:")
env.display()
```

```python
import random

class UtilitySecurityEnvironment:
    def __init__(self):
        self.components = {chr(i): random.choice(
            ["Safe", "Low", "High"]) for i in range(ord('A'), ord('J'))}

    def display(self):
        for c, s in self.components.items():
            print(f"{c}: {s}")


class UtilitySecurityAgent:
    def act(self, env):
        for c, s in env.components.items():
            if s == "Low":
                print(f" Patching LOW risk in {c}")
                env.components[c] = "Safe"
            elif s == "High":
                print(f"HIGH risk in {c} – Premium needed")
            else:
                print(f" {c} safe")


env = UtilitySecurityEnvironment()
agent = UtilitySecurityAgent()

print("\nInitial State:")
env.display()

agent.act(env)

print("\nFinal State:")
env.display()
```

```python
class HospitalEnvironment:
    def __init__(self):
        self.schedule = {
            "Room101": "Paracetamol",
            "Room102": "Insulin"
        }


class DeliveryRobotAgent:
    def achieve_goal(self, env):
        for room, med in env.schedule.items():
            self.move(room)
            self.scan_id(room)
            self.deliver(med)

    def move(self, room):
        print(f" Moving to {room}")

    def scan_id(self, room):
        print(f" Scanning patient ID in {room}")

    def deliver(self, med):
        print(f" Delivering {med}\n")


env = HospitalEnvironment()
agent = DeliveryRobotAgent()

agent.achieve_goal(env)
```

```python
class FireEnvironment:
    def __init__(self):
        self.grid = {
            'a': 'safe', 'b': 'safe', 'c': 'fire',
            'd': 'safe', 'e': 'fire', 'f': 'safe',
            'g': 'safe', 'h': 'safe', 'j': 'fire'
        }

    def display(self):
        symbols = {'safe': ' ', 'fire': 'F'}
        for i, r in enumerate(self.grid):
            print(symbols[self.grid[r]], end=" ")
            if (i + 1) % 3 == 0:
                print()


class FireRobotAgent:
    def __init__(self, path):
        self.path = path

    def act(self, env):
        for room in self.path:
            print(f"\n Entering {room}")
            if env.grid[room] == "fire":
                print("F Fire detected → Extigushing")
                env.grid[room] = "safe"
            env.display()


env = FireEnvironment()
agent = FireRobotAgent(['a','b','c','d','e','f','g','h','j'])

agent.act(env)
```
