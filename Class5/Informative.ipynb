```python
from queue import PriorityQueue

graph = { 
    'A': {'B': 2, 'C': 1},
    'B': {'D': 4, 'E': 3},
    'C': {'F': 1, 'G': 5},
    'D': {'H': 2},
    'E': {}, 
    'F': {'I': 6},
    'G': {},
    'H': {},
    'I': {}
}

class Node:
    
    def __init__(self, position, parent=None):
        self.position = position
        self.parent = parent
        self.cost = 0
        self.path_cost = 0

    def __lt__(self, other):
        return self.cost < other.cost


def best_first_search(start, end, graph):

    init_node = Node(start)

    frontier = PriorityQueue()
    visited = set()

    frontier.put(init_node)

    while not frontier.empty():

        current_node = frontier.get()
        current_position = current_node.position

        if current_position == end:

            path = []
            cost = current_node.path_cost

            while current_node:
                path.append(current_node.position)
                current_node = current_node.parent

            return (cost, path[::-1])

        visited.add(current_position)

        for neighbor, weight in graph[current_position].items():

            if neighbor not in visited:

                child = Node(neighbor, current_node)

                child.path_cost = current_node.path_cost + weight
                child.cost = child.path_cost

                frontier.put(child)

    return None


start_node = 'A'
end_node = 'I'

path = best_first_search(start_node, end_node, graph)

print("Path:", path)
```

```python
from queue import PriorityQueue

class Node:
    def __init__(self, position, parent=None):
        self.position = position
        self.parent = parent
        self.f = 0

    def __lt__(self, other):
        return self.f < other.f


def heuristic(a, goals):
    # Distance to the closest goal
    return min(abs(a[0]-g[0]) + abs(a[1]-g[1]) for g in goals)


def best_first_search(maze, start):

    rows = len(maze)
    cols = len(maze[0])

    # Automatically find all goal nodes marked with 2
    goals = [
        (r, c)
        for r in range(rows)
        for c in range(cols)
        if maze[r][c] == 2
    ]

    if not goals:
        print("No goal nodes found.")
        return None

    frontier = PriorityQueue()
    start_node = Node(start)
    start_node.f = heuristic(start, goals)
    frontier.put(start_node)

    visited = set()

    while not frontier.empty():

        current = frontier.get()

        if current.position in visited:
            continue

        visited.add(current.position)

        # Stop at ANY goal node
        if current.position in goals:
            path = []
            while current:
                path.append(current.position)
                current = current.parent
            print(path[::-1])
            return path[::-1]

        x, y = current.position

        for dx, dy in [(1,0),(-1,0),(0,1),(0,-1)]:

            nx = x + dx
            ny = y + dy

            if (0 <= nx < rows and
                0 <= ny < cols and
                maze[nx][ny] != 1 and        # ✅ walkable (0 or 2)
                (nx, ny) not in visited):

                new_node = Node((nx, ny), current)
                new_node.f = heuristic((nx, ny), goals)
                frontier.put(new_node)

    return None


maze = [
    [0, 0, 1, 0, 2],  # goal top-right
    [0, 0, 0, 0, 0],
    [0, 0, 1, 0, 1],
    [0, 2, 1, 0, 2],  # goal mid-right
    [0, 0, 0, 1, 0]
]

start = (0, 0)

path = best_first_search(maze, start)
print("Path to nearest goal:", path)
```
