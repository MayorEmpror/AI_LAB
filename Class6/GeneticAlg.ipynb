```python
from queue import PriorityQueue
from typing import Dict

heuristics: Dict[str, int]  = {
    'A': 7,
    'B': 6,
    'C': 2,
    'D': 1,
    'E': 0,
    'F': 3,
    'G': 0
}

graph: Dict[str, Dict[str, int]] = {
    'A': {'B': 1, 'C': 4},
    'B': {'D': 5, 'E': 12},
    'C': {'F': 2},
    'D': {'G': 3},
    'E': {'G': 1},
    'F': {'G': 2},
    'G': {}
}


class node:
    def __init__(self, position: str, cost:int, heu: int, parent=None):
        self.position = position
        self.parent = parent
        self.cost = cost
        self.heu = heu
        self.f  = 0

    def __lt__(self, other):
        return self.f < other.f


def a_star_search(graph, heu, start, end):
    start_node = node(start, 0, heuristics[start])
    start_node.f = heuristics[start]

    frontier = PriorityQueue()
    frontier.put(start_node)

    visited = set()
    best_g_cost = {start: 0}

    while not frontier.empty():

        current_node = frontier.get()
        current_position = current_node.position

        if current_position == end:

            path = []
            trace_node = current_node

            while trace_node:
                path.append(trace_node.position)
                trace_node = trace_node.parent

            return path[::-1]

        if current_position in visited:
            continue

        visited.add(current_position)

        for neighbours, edgecost in graph.get(current_position, {}).items():

            new_cost = edgecost + best_g_cost[current_position]
            if neighbours not in best_g_cost or new_cost < best_g_cost[neighbours]:

                best_g_cost[neighbours] = new_cost
                child = node(neighbours, new_cost, heuristics[neighbours], current_node)
                child.f = new_cost + heuristics[neighbours]

                frontier.put(child)

    return None


start = 'A'
end   = 'G'

path = a_star_search(graph, heuristics, start, end)
print("Response:", path)
```
