```python

class SecurityAgent:
    def __init__(self, agent_id, name, status="Active"):
        self.agent_id = agent_id
        self.name = name
        self.status = status

    def display_status(self):
        print(f"Agent ID: {self.agent_id}, Name: {self.name}, Status: {self.status}")



class FirewallAgent(SecurityAgent):
    def monitor_traffic(self):
        print(f"{self.name}: Monitoring network traffic for suspicious activity.")



class MalwareDetectionAgent(SecurityAgent):
    def scan_files(self):
        print(f"{self.name}: Scaning files for malware and threats.")



class AutomationAgent(SecurityAgent):
    def run_automation(self):
        print(f"{self.name}: Runing AI-based automation to optimie system security.")



firewall = FirewallAgent(101, "Firewall Defender")
malware_detector = MalwareDetectionAgent(102, "Malware Scanner")
automation = AutomationAgent(103, "Security Optimizer")


firewall.display_status()
malware_detector.display_status()
automation.display_status()

print("\nAI Cybersecurity System in Action")


firewall.monitor_traffic()
malware_detector.scan_files()
automation.run_automation()
```

```python
class Threats:
    def __init__(self, threat_id, name, severity):
                self._threat_id = threat_id
                self._name = name
                self._severity = severity

class PhishingThreat(Threats):
    def __init__(self, email, name, threat_id, severity):
                super().__init__(threat_id, name, severity)
                self._email = email
    def analyze_email(self):
        print("monitoring emails")
    

class RansomwareThreat(Threats):
    def __init__(self, files, name, threat_id, severity):
                super().__init__(threat_id, name, severity)
                self._files = files

    def scan_files(self):
                print("Monitoring files")

class BotnetThreat(Threats):
    def __init__(self, traffic, name, threat_id, severity):
                super().__init__(threat_id, name, severity)
                self._traffic = traffic
    
    def detect_traffic(self):
                print("monitoring traffic")
                

email = "lorem ipsum dolor ammet consetct"
file = "lorem ipsum dolor ammet consetct"
name = "John Doe"
traffic = "lorem ipsum"


Phishingthreat = PhishingThreat(email, name, 1, 80)
Ransomwarethreat = RansomwareThreat(file, name, 2, 60)
Botnetthreat = BotnetThreat(traffic, name, 3, 45)

Phishingthreat.analyze_email()
Ransomwarethreat.scan_files()
Botnetthreat.detect_traffic()



    



```

```python

class ResponseAgent:
    def execute_response(self):
        pass


class AlertAgent(ResponseAgent):
    def __init__(self, notification):
        self.notification = notification
                
    def execute_response(self):
        print(f"message: {self.notification}")


class BlockAgent(ResponseAgent):
    def execute_response(self):
        print("blocked malicious threats")


class RecoverAgent(ResponseAgent):
    def execute_response(self):
        print("recovered lost data")



agents = [
    AlertAgent("Your account has been hacked"),
    BlockAgent(),
    RecoverAgent()
]


for agent in agents:
    agent.execute_response()
```

```python

class Employee:
    def work(self):
        print("Employee is working.")



class Manager(Employee):
    def work(self):
        print("Manager is planning projects and managing the team.")


class Developer(Employee):
    def work(self):
        print("Developer is writing and debugging code.")


class Designer(Employee):
    def work(self):
        print("Designer is creating UI/UX designs.")



employees = [
    Manager(),
    Developer(),
    Designer()
]


for employee in employees:
    employee.work()
```

```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.__grade = grade   

 
    def set_grade(self, grade):
        if 0 <= grade <= 100:
            self.__grade = grade
        else:
            print("Inalid grade! Grad must be between 0 and 100.")

  
    def get_grade(self):
        return self.__grade


    def display_info(self):
        print(f"Name: {self.name}, Grad: {self.__grade}")



student1 = Student("Alice", 85)
student2 = Student("Bob", 90)


student1.set_grade(92)
student2.set_grade(88)


print("Alice's Grad:", student1.get_grade())
print("Bob's Grad:", student2.get_grade())


student1.display_info()
student2.display_info()

```
