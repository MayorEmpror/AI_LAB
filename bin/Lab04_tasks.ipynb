```python
# Q1(A)
class goalbasedagent:
    def __init__(self,goal):
        self.goal = goal 

    def forumlate_goal(self, percept):
        if percept == self.goal:
            return "goal reached"
        return "searching"
    

    def dfs_lim(self, graph, start, goal, depth_limit):
     visited = []
     def dfs(node, depth):
        if depth > depth_limit:
            return None
        visited.append(node)
        if node == goal:
            print(f"Goal found at {node}")
        for neighbor in graph.get(node, []):
            if neighbor not in visited:
                path = dfs(neighbor, depth+ 1)
                if path :
                    return path
        visited.pop()
        return None 
     return dfs(start, 0)
    
    
    def act(self, percept, graph, limit):
        goal_status = self.forumlate_goal(percept)
        if goal_status == "goal reached":
            return f"goal {self.goal} found !"
        else:
            return self.dfs_lim(graph, percept,self.goal,limit)
    
class Enviornment:
    def __init__(self, graph):
        self.graph = graph
    def get_percept(self, node):
        return node

def run_agent(agent, env, start_node, depth_limit):
    percept = env.get_percept(start_node)
    action = agent.act(percept, env.graph, depth_limit)
    print("Result:", action)


tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': [],
    'F': ['I'],
    'G': [],
    'H': [],
    'I': []
}
start_node = 'A'
goal_node = 'I'

agent = goalbasedagent(goal_node)
env = Enviornment(tree)
run_agent(agent, env, start_node,4)

```

```python
# Q1(B)
class goalbasedagent:
    def __init__(self, goal):
        self.goal = goal 

    def formulate_goal(self, percept):
        if percept == self.goal:
            return "goal achived"
        else :
            return "serching"
    

    def UCS(self,graph, start):
        frontier = [(start, 0)]
        visited = set()
        Total_cost = {start : 0}
        prev_node = {start: None}

        while frontier:
            frontier.sort(key=lambda x : x[1] )
            current_node, current_cost = frontier.pop(0)
            if current_node in visited:
                continue 
            visited.add(current_node)

           
            if current_node == self.goal:
                path = []
                while current_node is not None:
                    path.append(current_node)
                    current_node = prev_node[current_node]
                path.reverse()
                print(f"goal found with UCS, Path :{path}, total cost : {current_cost}")
                return


            for neighbor, cost in graph[current_node].items():
               new_cost =  current_cost + cost 
               if neighbor not in Total_cost or new_cost < Total_cost[neighbor]:
                    Total_cost[neighbor] = new_cost
                    prev_node[neighbor] = current_node
                    frontier.append((neighbor, new_cost))
        print("goal not found")


    def act(self, graph, start, percept):
        goal_status =  self.formulate_goal(percept)
        if goal_status == "Goal reached":
            return f"Goal {self.goal} found"
        else:
            return self.UCS(graph, start)

class Enviornment:
    def __init__(self, graph):
        self.graph = graph
    
    def get_percept(self, node):
        return node 

def run_agent(agent, enviornment, start_node):
    percept = enviornment.get_percept(start_node)
    action = agent.act(enviornment.graph, start_node,percept)
    print(action)

graph = {
    'A':{'B': 2, 'C': 1} ,
    'B': {'D': 4, 'E': 3},
    'C': {'F': 1, 'G': 5},
    'D': {'H': 2},
    'E': {},
    'F': {'I': 6},
    'G': {},
    'H': {},
    'I': {}
}


start_node = 'A'
goal_node = 'I'

agent = goalbasedagent(goal_node)
env = Enviornment(graph)

run_agent(agent, env, start_node)



```

```python
# Q2
def tsp(cost):
    n = len(cost)
    
    visited = [False] * n
    visited[0] = True  
    
    min_cost = float('inf')
    best_path = []

    def backtrack(current_city, count, current_cost, path):
        nonlocal min_cost, best_path
        
   
        if count == n:
            total_cost = current_cost + cost[current_city][0]  
            
            if total_cost < min_cost:
                min_cost = total_cost
                best_path = path + [0]
            return
        
      
        for next_city in range(n):
            if not visited[next_city]:
                visited[next_city] = True
                
                backtrack(
                    next_city,
                    count + 1,
                    current_cost + cost[current_city][next_city],
                    path + [next_city]
                )
                
                visited[next_city] = False  

    backtrack(0, 1, 0, [0])
    
    return best_path, min_cost



cost = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

path, min_cost = tsp(cost)


path_display = [city + 1 for city in path]

print("Shortest Route:", path_display)
print("Minimum Cost:", min_cost)
```

```python
# Q3
def dls(graph, current, goal, depth, visited, path):

    path.append(current)
    visited.add(current)


    if current == goal:
        return True


    if depth == 0:
        path.pop()
        return False

    
    for neighbor in graph.get(current, []):
        if neighbor not in visited:
            if dls(graph, neighbor, goal, depth - 1, visited, path):
                return True

   
    path.pop()
    return False



def iddfs(graph, start, goal, max_depth):
    for depth in range(max_depth + 1):
        visited = set()
        path = []
        
        print(f"Searching at depth: {depth}")
        
        if dls(graph, start, goal, depth, visited, path):
            return path 
    
    return None  



graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}


result = iddfs(graph, 'A', 'F', 4)

if result:
    print("Goal found!")
    print("Path:", result)
else:
    print("Goal not found within depth limit.")
```
