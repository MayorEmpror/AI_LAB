```python
from ortools.sat.python import  cp_model  

model = cp_model.CpModel()

A = model.NewIntVar(1, 3, 'Red')
B = model.NewIntVar(1, 3, 'Blue')
C = model.NewIntVar(1, 3, 'Green')

model.Add(A != B)
model.Add(B != C)
model.Add(A != C)

solver = cp_model.CpSolver()
solution_printer = cp_model.VarArraySolutionPrinter([A, B, C])
solver.parameters.enumerate_all_solutions =  True
solver.Solve(model, solution_printer)
```

```python
from ortools.sat.python import cp_model 
model = cp_model.CpModel()

Alice = model.NewIntVar(0, 2, 'Alice')
Bob  = model.NewIntVar(0, 2, 'Bob')
Charlie = model.NewIntVar(0, 2, 'Charlie')

model.Add(Alice != Bob)

solver = cp_model.CpSolver()
solution_printer = cp_model.VarArraySolutionPrinter([Alice, Bob, Charlie])
solver.parameters.enumerate_all_solutions = True
solver.Solve(model, solution_printer)

```
