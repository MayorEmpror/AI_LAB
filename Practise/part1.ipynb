```python
import random
import math

cities = [(2,3),(5,4),(1,8),(7,2),(6,6),
          (8,3),(3,7),(9,5),(4,1),(2,6)]

NUM = len(cities)

def dist(a,b):
    return math.sqrt((a[0]-b[0])**2+(a[1]-b[1])**2)

def total(route):
    d = 0
    for i in range(NUM-1):
        d += dist(cities[route[i]],cities[route[i+1]])
    d += dist(cities[route[-1]],cities[route[0]])
    return d

# make random routes
pop = []
for i in range(50):
    r = list(range(NUM))
    random.shuffle(r)
    pop.append(r)

# GA loop
for g in range(200):

    pop.sort(key=total)

    new_pop = pop[:10]   # keep best

    while len(new_pop) < 50:

        p1 = random.choice(pop[:20])
        p2 = random.choice(pop[:20])

        cut = random.randint(1,NUM-2)

        child = p1[:cut]

        for c in p2:
            if c not in child:
                child.append(c)

        # simple mutation
        if random.random()<0.1:
            i = random.randint(0,NUM-1)
            j = random.randint(0,NUM-1)
            child[i],child[j]=child[j],child[i]

        new_pop.append(child)

    pop = new_pop

best = pop[0]

print(best)
print(total(best))
```

```python
import random

class Agent:
    def __init__(self):
        self.Q = {}
        self.action = ['Clean', 'Move']
        self.alpha = 0.1
        self.gamma = 0.9
        self.epsilon = 0.1

    def getQ(self, s, a):
        return self.Q.get((s, a), 0)

    def choose(self, s):
        if random.random() < self.epsilon:
            return random.choice(self.action)
        return max(self.action, key=lambda a: self.getQ(s, a))

    def learn(self, s, a, r, s2):
        old = self.getQ(s, a)
        future = max(self.getQ(s2, x) for x in self.action)
        self.Q[(s, a)] = old + self.alpha * (r + self.gamma * future - old)


# 🌍 Simple Environment
class Env:
    def __init__(self):
        self.rooms = {'A': random.choice(['Dirty', 'Clean']),
                      'B': random.choice(['Dirty', 'Clean'])}
        self.loc = 'A'

    def get_state(self):
        return (self.loc, self.rooms[self.loc])

    def step(self, action):
        if action == 'Clean':
            if self.rooms[self.loc] == 'Dirty':
                self.rooms[self.loc] = 'Clean'
                return 10   # reward
            else:
                return -1   # waste

        if action == 'Move':
            self.loc = 'B' if self.loc == 'A' else 'A'
            return -1       # movement cost


# 🚀 Run
agent = Agent()
env = Env()

for i in range(20):
    s = env.get_state()
    a = agent.choose(s)
    r = env.step(a)
    s2 = env.get_state()
    agent.learn(s, a, r, s2)

    print(s, a, r)
```

```python
import random

class leariningAgent:
    def __init__(self):
        self.Q = {}
        self.action = ['Clean', 'Move']
        self.alhpa = 0.1
        self.gamma = 0.9
        self.elipson = 0.3

    def getQ(self, state, action):
        return self.Q.get((state, action), 0.0)

    def choice(self, state):
        if random.random() < self.elipson:
            return random.choice(self.action)
        else:
            return max(self.action, key=lambda a: self.getQ(state, a))

    def learn(self, state, action, reward, next_state):
        old = self.getQ(state, action)
        future = max(self.getQ(next_state, x) for x in self.action)
        self.Q[(state, action)] = old + self.alhpa * (reward + self.gamma * future - old)


class env:
    def __init__(self):
        self.rooms = {
            'A': random.choice(['Dirty', 'Clean']),
            'B': random.choice(['Dirty', 'Clean'])
        }
        self.loc = 'A'

    def get_state(self):
        return (self.loc, self.rooms[self.loc])

    def act(self, action):
        if action == 'Clean':
            if self.rooms[self.loc] == 'Dirty':
                self.rooms[self.loc] = 'Clean'
                return 10
            else:
                return -1

        if action == 'Move':
            self.loc = 'B' if self.loc == 'A' else 'A'
            return -1


agent = leariningAgent()
envi = env()

for i in range(20):
    s = envi.get_state()
    a = agent.choice(s)
    r = envi.act(a)
    s2 = envi.get_state()
    agent.learn(s, a, r, s2)

    print(s, a, r)
```
