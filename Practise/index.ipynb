```python
import queue 

tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': [],
    'F': ['I'],
    'G': [],
    'H': [],
    'I': []
}

class node :
    def __init__(self, position, parent=None):
        self.position = position
        self.parent = parent


def breadth_first_search(tree, start, goal):
        frontier = queue.Queue()
        start_node =  node(start)
        frontier.put(start_node)

        while not frontier.empty():
            current_node = frontier.get()
            if current_node.position == goal :
                path = []
                temp = current_node 
                while temp:
                    path.append(temp.position)
                    temp = temp.parent
                return path[::-1]
            
            for neighbours in tree.get(current_node.position, []):
                new_node = node(neighbours, current_node)
                frontier.put(new_node)
        return None



start_node = 'A'
end_node = 'I'

path = breadth_first_search(tree, start_node, end_node)
print(path)




```

```python
from queue import PriorityQueue

graph = {
    'A': {'B': 5, 'C': 2},
    'B': {'D': 4, 'E': 2},
    'C': {'B': 8, 'E': 7},
    'D': {'E': 6, 'F': 3},
    'E': {'F': 1},
    'F': {}
}

class Node:
    def __init__(self, position, parent=None, cost=0):
        self.position = position
        self.parent = parent
        self.cost = cost

    def __lt__(self, other):
        return self.cost < other.cost  # ensures PriorityQueue pops lowest-cost node

def UCS(tree, start, goal):
    start_node = Node(start, cost=0)
    frontier = PriorityQueue()
    frontier.put(start_node)

    visited = set()
    cost_till_now = {start: 0}

    while not frontier.empty():
        current_node = frontier.get()

        # Only mark visited when popping
        if current_node.position in visited:
            continue
        visited.add(current_node.position)

        # Goal check
        if current_node.position == goal:
            path = []
            temp = current_node
            while temp:
                path.append(temp.position)
                temp = temp.parent
            return path[::-1]

        # Add neighbors
        for neighbour, edge_cost in tree.get(current_node.position, {}).items():
            new_cost = current_node.cost + edge_cost
            if neighbour not in cost_till_now or new_cost < cost_till_now[neighbour]:
                cost_till_now[neighbour] = new_cost
                new_node = Node(neighbour, current_node, new_cost)
                frontier.put(new_node)

    return None

# Run UCS
start = 'A'
end = 'F'
path = UCS(graph, start, end)
print("Path:", path)
```

```python



class utilitybased_agent:
    def __init__(self, size):
        self.size = size
        self.model = [['?' for i in range(size)] for j in range(size)]
        self.utility = 0

    def update_model(self, dx, dy, percept):
        self.model[dx][dy] = percept

    def calculate_utility(self, percept):

        if percept == 'D':
            return 10   # reward for cleaning

        return -1       # small movement cost

    def action(self, dx, dy, percept):

        self.update_model(dx, dy, percept)

        score = self.calculate_utility(percept)

        if percept == 'D':
            self.utility += score
            return "CLEAN"

        self.utility += score
        return "MOVE"

class environment:
    def __init__(self):
        self.grid = [
            ['C','D','C'],
            ['C','C','C'],
            ['D','D','C'],
        ]

    def get_percept(self, dx, dy):
        return self.grid[dx][dy]




def run_agent(environment, agent, grid_size):

    for dx in range(grid_size):
        for dy in range(grid_size):

            percept = environment.get_percept(dx, dy)

            action = agent.action(dx, dy, percept)

            if action == "CLEAN":
                environment.grid[dx][dy] = 'C'
                agent.update_model(dx, dy, 'C')
                print(f"Cleaned at {dx},{dy}")

    print("Final Environment:")
    for row in environment.grid:
        print(row)

    print("Total Utility:",agent.utility)

env = environment()
agent = utilitybased_agent(3)

run_agent(env, agent, 3)
```

```python
import heapq

# ---------- ENVIRONMENT ----------
class Environment:
    def __init__(self):
        self.grid = [
            ['C','D','C'],
            ['C','C','C'],
            ['D','D','C']
        ]
        self.size = 3

    def dirty_cells(self):
        return [(i,j) for i in range(self.size) for j in range(self.size) if self.grid[i][j]=='D']

    def clean(self, x, y):
        self.grid[x][y] = 'C'

    def show(self):
        for row in self.grid:
            print(row)
        print()

# ---------- UTILITY AGENT WITH A* ----------
class UtilityAgent:
    def __init__(self, start=(0,0)):
        self.pos = start
        self.utility = 0

    def neighbors(self, x, y):
        moves = [(x-1,y),(x+1,y),(x,y-1),(x,y+1)]
        return [(nx,ny) for nx,ny in moves if 0<=nx<3 and 0<=ny<3]

    def heuristic(self, a,b):
        return abs(a[0]-b[0])+abs(a[1]-b[1])

    def astar(self, target):
        start = self.pos
        frontier = [(0,start)]
        came_from = {start:None}
        cost = {start:0}

        while frontier:
            _, current = heapq.heappop(frontier)
            if current==target: break
            for nxt in self.neighbors(*current):
                new_cost = cost[current]+1
                if nxt not in cost or new_cost<cost[nxt]:
                    cost[nxt]=new_cost
                    heapq.heappush(frontier,(new_cost+self.heuristic(nxt,target),nxt))
                    came_from[nxt]=current

        path=[]
        node=target
        while node!=start:
            path.append(node)
            node=came_from[node]
        path.reverse()
        return path

    def act(self, env):
        x,y = self.pos
        if env.grid[x][y]=='D':
            env.clean(x,y)
            self.utility += 10
            print(f"Cleaned {self.pos}, utility={self.utility}")
            return "CLEAN"

        dirt = env.dirty_cells()
        if not dirt:
            return "STOP"

        nearest = min(dirt, key=lambda d: len(self.astar(d)))
        path = self.astar(nearest)
        if path:
            self.pos = path[0]
            self.utility -= 1
            print(f"Moved to {self.pos}, utility={self.utility}")
        return "MOVE"

# ---------- RUN ----------
env = Environment()
agent = UtilityAgent()

while True:
    action = agent.act(env)
    if action=="STOP":
        break

print("\nFinal Environment:")
env.show()
print("Total Utility:", agent.utility)
```

```python
from ortools.sat.python import cp_model

def tsp_example():
    # Example distance matrix for 10 cities (symmetric)
    distance = [
        [0, 29, 20, 21, 16, 31, 100, 12, 4, 31],
        [29, 0, 15, 29, 28, 40, 72, 21, 29, 41],
        [20, 15, 0, 15, 14, 25, 81, 9, 23, 27],
        [21, 29, 15, 0, 4, 12, 92, 12, 25, 13],
        [16, 28, 14, 4, 0, 16, 94, 9, 20, 16],
        [31, 40, 25, 12, 16, 0, 95, 24, 36, 3],
        [100,72,81,92,94,95,0,90,101,99],
        [12,21,9,12,9,24,90,0,15,25],
        [4,29,23,25,20,36,101,15,0,35],
        [31,41,27,13,16,3,99,25,35,0]
    ]
    
    num_cities = len(distance)
    
    model = cp_model.CpModel()
    
    # Variables: position[i] = city at position i
    positions = [model.NewIntVar(0, num_cities-1, f'pos_{i}') for i in range(num_cities)]
    
    # Constraint: all cities visited exactly once
    model.AddAllDifferent(positions)
    
    # Helper: calculate total distance
    total_distance = model.NewIntVar(0, 10000, 'total_distance')
    
    # Sum distances between consecutive cities including return to start
    distance_terms = []
    for i in range(num_cities):
        next_i = (i+1) % num_cities
        # Create intermediate distance variable for this step
        d = model.NewIntVar(0, 1000, f'd_{i}')
        model.AddElement(positions[i]*num_cities + positions[next_i], 
                         [distance[r][c] for r in range(num_cities) for c in range(num_cities)],
                         d)
        distance_terms.append(d)
    
    model.Add(total_distance == sum(distance_terms))
    
    # Objective: minimize total distance
    model.Minimize(total_distance)
    
    # Solve
    solver = cp_model.CpSolver()
    status = solver.Solve(model)
    
    if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
        route = [solver.Value(pos) for pos in positions]
        print("Optimal route:", route)
        print("Total distance:", solver.Value(total_distance))
    else:
        print("No solution found.")

tsp_example()
```

```python
from ortools.sat.python import cp_model

# ---------------- GRID ----------------
grid = [
    ['S','Z','Z','Z','Z'],
    ['Z','O','Z','Z','Z'],
    ['Z','Z','Z','O','Z'],
    ['Z','O','Z','Z','Z'],
    ['Z','Z','Z','Z','T']
]

n = len(grid)
start = (0,0)
target = (4,4)
max_steps = 8  # max number of moves (diagonals)

# ---------------- MODEL ----------------
model = cp_model.CpModel()

# Positions variables
positions = [
    (model.NewIntVar(0,n-1,f"x_{i}"), model.NewIntVar(0,n-1,f"y_{i}"))
    for i in range(max_steps)
]

# ---------------- CONSTRAINTS ----------------
# Start and end
model.Add(positions[0][0] == start[0])
model.Add(positions[0][1] == start[1])
model.Add(positions[-1][0] == target[0])
model.Add(positions[-1][1] == target[1])

# Diagonal moves & obstacle avoidance
for i in range(max_steps - 1):
    x1, y1 = positions[i]
    x2, y2 = positions[i+1]

    # Diagonal: |dx| == 1 and |dy| == 1
    dx = model.NewIntVar(-1,1,f"dx_{i}")
    dy = model.NewIntVar(-1,1,f"dy_{i}")
    model.Add(dx == x2 - x1)
    model.Add(dy == y2 - y1)
    model.AddAbsEquality(model.NewIntVar(1,1,f"d_abs_{i}"), dx)
    model.AddAbsEquality(model.NewIntVar(1,1,f"d_abs2_{i}"), dy)

    # Avoid obstacles
    for r in range(n):
        for c in range(n):
            if grid[r][c] == 'O':
                model.AddForbiddenAssignments([x2,y2], [(r,c)])

# ---------------- SOLVER ----------------
class Printer(cp_model.CpSolverSolutionCallback):
    def OnSolutionCallback(self):
        path = [(self.Value(x), self.Value(y)) for x,y in positions]
        print("Path:", path)


solver = cp_model.CpSolver()
status = solver.Solve(model, Printer())



if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
    print("Solution found!")
else:
    print("No solution exists.")
```
