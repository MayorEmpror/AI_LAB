```python

s = input("Enter a string: ")

longest_substring = ""
current_substring = ""


for ch in s:

    if ch in current_substring:

        index = 0
        while current_substring[index] != ch:
            index += 1
        current_substring = current_substring[index + 1:]

   
    current_substring += ch

   
    if len(current_substring) > len(longest_substring):
        longest_substring = current_substring


print("Longest substring without repeating characters:", longest_substring)
```

```python
def find_repeated_numbers(numbers):
    repeated = []
    

    for i in range(len(numbers)):
        count = 0
        
       
        for j in range(len(numbers)):
            if numbers[i] == numbers[j]:
                count += 1
        
      
        if count > 1 and numbers[i] not in repeated:
            repeated.append(numbers[i])
    
    return repeated



nums = [1, 2, 3, 2, 4, 5, 1, 3, 6]
result = find_repeated_numbers(nums)

print("Repeated numbers:", result)
```
